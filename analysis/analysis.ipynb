```python
 from read_and_merge_adds import get_adds

ads = get_adds().values()
            
```

```python
from matplotlib import pyplot as plt

def type_check(add):
    return add["type"][0] == "realestate"
        

filtered_ads = list(filter(type_check, ads))


for ad in filtered_ads:
    
    plt.plot( range(len(ad["timestamp"])), list(map(lambda ad: ad["amount"], ad["price_total"])), label=ad["ad_id"][0])

plt.show()
    
```
